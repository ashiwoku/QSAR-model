```python
from sklearn.feature_selection import VarianceThreshold
```

```python
threshold = VarianceThreshold(threshold=.1)
```

```python
labels = classification_data.class_label
```

```python
classification_data.drop(columns='class_label', inplace=True)
```

```python
def variance_threshold_selector(data, threshold=0.15):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    return data[data.columns[selector.get_support(indices=True)]]
```

```python
classification_data_HV = variance_threshold_selector(classification_data)
```

```python
classification_data_HV = pd.merge(classification_data_HV,labels,left_index=True,right_index=True)
```

```python
classification_data_HV ## removing features that have a variance less than 15%
```

```python
regression_data_HV = variance_threshold_selector(regression_data)
```

```python
regression_data_HV
```

```python
## remove features that display high correlation with many othe features 
```

```python
correlated_features_1 = set()
corr_matrix_reg = regression_data_HV.corr() 
```

```python
corr_matrix_reg
```

```python
for i in range(len(corr_matrix_reg.columns)):
    for j in range(i):
        if abs(corr_matrix_reg.iloc[i, j]) > 0.8:
            colname = corr_matrix_reg.columns[i]
            correlated_features_1.add(colname)
```

```python
correlated_features_1
```

```python
correlated_features_2 = set()
corr_matrix_class = classification_data_HV.corr() 
```

```python
for i in range(len(corr_matrix_class.columns)):
    for j in range(i):
        if abs(corr_matrix_class.iloc[i, j]) > 0.8:
            colname = corr_matrix_class.columns[i]
            correlated_features_2.add(colname)
```

```python
correlated_features_2
```

```python
## Remove correlated features.
def remove_correlated_features(features, data):
    for x in features:
        data.drop(x, axis=1, inplace=True)
    return data
```

```python
regression_data_HV = remove_correlated_features(correlated_features_1,regression_data_HV)
```

```python
regression_data_HV
```

```python
classification_data_HV = remove_correlated_features(correlated_features_2, classification_data_HV)
```

```python
classification_data_HV
```

```python
regression_data_HV.to_csv('regression_stepwise_ready.csv', index=False)
classification_data_HV.to_csv('classification_stepwise_ready.csv', index=False)
```
